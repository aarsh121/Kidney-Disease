```python
import pandas as pd
```

```python
df = pd.read_csv('ckd.csv')
```

```python
df
```

```python
X_new = df.drop(['class'], axis=1)
y=df['class']
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
for i in df:
  print("{} Max: {} Min: {}".format(i, df[i].max(),df[i].min()))
```

```python
wbcc = ctrl.Antecedent(np.arange(2200, 26400, 1), 'white blood cell count')
bgr = ctrl.Antecedent(np.arange(22, 490, 1), 'blood glucose')
bu = ctrl.Antecedent(np.arange(1.5, 391, 1), 'blood urea')
sc = ctrl.Antecedent(np.arange(0.4, 76, 1), 'serum creatinine')
pcv = ctrl.Antecedent(np.arange(9, 54, 1), 'packed cell volume')
al = ctrl.Antecedent(np.arange(0, 5, 1), 'albumin')
hem = ctrl.Antecedent(np.arange(3.1, 17.8, 1), 'hemoglobin')
age = ctrl.Antecedent(np.arange(2, 90, 1), 'age')
sugar = ctrl.Antecedent(np.arange(0, 5, 1), 'sugar')
hy = ctrl.Antecedent(np.arange(0, 1, 1), 'hypertension')
```

```python
wbcc.automf(3)
bgr.automf(3)
bu.automf(3)
sc.automf(3)
pcv.automf(3)
al.automf(3)
hem.automf(3)
age.automf(3)
sugar.automf(3)
hy.automf(3)
```

```python
def max_and_fuzzy(p,a,g):
  fuzzy_val = 0
  if (p >= a) and (p >= g):
    fuzzy_val = 0
  elif (a >= p) and (a >= g):
    fuzzy_val = 1
  else:
    fuzzy_val = 2
  return fuzzy_val
```

```python
new_val = []
for i in X_new.itertuples():
  val = []
  p = fuzz.interp_membership(wbcc.universe, wbcc['poor'].mf, i[1])
  a = fuzz.interp_membership(wbcc.universe, wbcc['average'].mf, i[1])
  g = fuzz.interp_membership(wbcc.universe, wbcc['good'].mf, i[1])
  val.append(max_and_fuzzy(p,a,g))

  p = fuzz.interp_membership(bgr.universe, bgr['poor'].mf, i[2])
  a = fuzz.interp_membership(bgr.universe, bgr['average'].mf, i[2])
  g = fuzz.interp_membership(bgr.universe, bgr['good'].mf, i[2])
  val.append(max_and_fuzzy(p,a,g))

  p = fuzz.interp_membership(bu.universe, bu['poor'].mf, i[3])
  a = fuzz.interp_membership(bu.universe, bu['average'].mf, i[3])
  g = fuzz.interp_membership(bu.universe, bu['good'].mf, i[3])
  val.append(max_and_fuzzy(p,a,g))

  p = fuzz.interp_membership(sc.universe, sc['poor'].mf, i[4])
  a = fuzz.interp_membership(sc.universe, sc['average'].mf, i[4])
  g = fuzz.interp_membership(sc.universe, sc['good'].mf, i[4])
  val.append(max_and_fuzzy(p,a,g))

  p = fuzz.interp_membership(pcv.universe, pcv['poor'].mf, i[5])
  a = fuzz.interp_membership(pcv.universe, pcv['average'].mf, i[5])
  g = fuzz.interp_membership(pcv.universe, pcv['good'].mf, i[5])
  val.append(max_and_fuzzy(p,a,g))

  p = fuzz.interp_membership(al.universe, al['poor'].mf, i[6])
  a = fuzz.interp_membership(al.universe, al['average'].mf, i[6])
  g = fuzz.interp_membership(al.universe, al['good'].mf, i[6])
  val.append(max_and_fuzzy(p,a,g))

  p = fuzz.interp_membership(hem.universe, hem['poor'].mf, i[7])
  a = fuzz.interp_membership(hem.universe, hem['average'].mf, i[7])
  g = fuzz.interp_membership(hem.universe, hem['good'].mf, i[7])
  val.append(max_and_fuzzy(p,a,g))

  p = fuzz.interp_membership(age.universe, age['poor'].mf, i[8])
  a = fuzz.interp_membership(age.universe, age['average'].mf, i[8])
  g = fuzz.interp_membership(age.universe, age['good'].mf, i[8])
  val.append(max_and_fuzzy(p,a,g))


  p = fuzz.interp_membership(sugar.universe, sugar['poor'].mf, i[9])
  a = fuzz.interp_membership(sugar.universe, sugar['average'].mf, i[9])
  g = fuzz.interp_membership(sugar.universe, sugar['good'].mf, i[9])
  val.append(max_and_fuzzy(p,a,g))


  p = fuzz.interp_membership(hy.universe, hy['poor'].mf, i[10])
  a = fuzz.interp_membership(hy.universe, hy['average'].mf, i[10])
  g = fuzz.interp_membership(hy.universe, hy['good'].mf, i[10])
  val.append(max_and_fuzzy(p,a,g))
  new_val.append(val)
```

```python
new_val
```

```python
selected_col = df.drop(['class'], axis=1)
```

```python
c_l = []
for val in selected_col:
  c_l.append(val)
```

```python
df_l = pd.DataFrame(new_val, columns=c_l)
```

```python
df_l
```

```python
copy_df = df_l
```

```python
copy_df['class'] = y
```

```python
copy_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(df_l,y,train_size=0.75)
```

```python
from sklearn import svm
lin_clf = svm.LinearSVC()
lin_clf.fit(X_train, y_train)
ypred = lin_clf.predict(X_test)
print(accuracy_score(ypred,y_test))
```
